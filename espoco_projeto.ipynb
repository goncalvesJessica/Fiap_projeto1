```python


pasta = r"C:\Users\jessi\Pos FISP\Desafios Pos\Projeto_1\dados.db"

# Ler só colunas necessárias + converter data
order_items = pd.read_csv(
    f"{pasta}/olist_order_items_dataset.csv",
    usecols=["product_id", "shipping_limit_date"],
    parse_dates=["shipping_limit_date"]
)

# Filtrar período (NOVEMBRO 2017)
order_items = order_items[
    (order_items["shipping_limit_date"] >= "2017-11-01") &
    (order_items["shipping_limit_date"] <= "2017-11-30")
]

# Ler produtos já enxuto
products = pd.read_csv(
    f"{pasta}/olist_products_dataset.csv",
    usecols=["product_id", "product_category_name", "product_name_lenght"]
)

# Filtrar móveis antes do merge
products_moveis = products.loc[
    products["product_category_name"] == "moveis_decoracao",
    ["product_id", "product_name_lenght"]
]

# Merge
df = order_items.merge(
    products_moveis,
    on="product_id",
    how="inner"
)

# Agrupamento mais performático
resultado = (
    df.groupby("product_name_lenght")
    .size()
    .sort_index()
)

print(resultado)
```

```python
#Valido quantos itens de móveis foram vendidos por ano, e quantos pedidos únicos tiveram pelo menos um móvel.
pasta = r"C:\Users\jessi\Pos FISP\Desafios Pos\Projeto_1\dados.db"

orders = pd.read_csv(
    f"{pasta}/olist_orders_dataset.csv",
    usecols=["order_id", "order_purchase_timestamp"],
    parse_dates=["order_purchase_timestamp"]
)

# Criar ANO do pedido
orders["ano"] = orders["order_purchase_timestamp"].dt.year

order_items = pd.read_csv(
    f"{pasta}/olist_order_items_dataset.csv",
    usecols=["order_id", "product_id"]
)

products = pd.read_csv(
    f"{pasta}/olist_products_dataset.csv",
    usecols=["product_id", "product_category_name"]
)

# Filtrar móveis pra deixa mais leve
products_moveis = products.loc[
    products["product_category_name"] == "cama_mesa_banho",
    ["product_id"]
]

# Itens + produtos
df = order_items.merge(
    products_moveis,
    on="product_id",
    how="inner"
)

# Adicionar dados do pedido
df = df.merge(
    orders,
    on="order_id",
    how="inner"
)

resultado = df.groupby("ano").agg(
    pedidos=("order_id", "nunique"),   
    itens=("order_id", "size")         
).sort_index()

print(resultado)
```

```python
import matplotlib.pyplot as plt

# Top 15 mais frequentes
resultado = df["product_name_lenght"].value_counts().head(15)

# Ordenar para o gráfico ficar mais bonito (crescente)
resultado = resultado.sort_index()

# Plot
resultado.plot(kind="bar", figsize=(10,6))

plt.title("Top 15 Tamanhos de Nome de Produto (Mais Frequentes)")
plt.xlabel("Quantidade de caracteres no nome")
plt.ylabel("Quantidade de pedidos")

plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
